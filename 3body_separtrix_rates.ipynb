# Calculation of 3 body recombination for finite velocity and mass ion

```python
import matplotlib.pyplot as plt
from numpy import *
import scipy
from scipy.integrate import quad, dblquad, tplquad

import warnings
warnings.filterwarnings("ignore")
```

```python
def r_bar(xi, eta):
    value = minimum(1.0, exp( xi**(-0.5) * eta**(-1.5)) * xi**(-1.125) * eta**(3.375))
    return value
```

```python
xi_min = -0.5
xi_max = 3.0
xi_num = 8

eta_min = -2.0
eta_max = 1.0
eta_num = 50

xi_values = linspace(xi_min, xi_max, num=xi_num)
eta_values = linspace(eta_min, eta_max, num=eta_num)
r_bar_values = zeros((xi_num, eta_num))

for i, xi in enumerate(xi_values):
    for j, eta in enumerate(eta_values):
        r_bar_values[i,j] = r_bar(10**xi, 10**eta)
```

```python
fig, ax = plt.subplots(figsize=[6.0,6.0])
for i, xi in enumerate(xi_values):
    plt.plot(eta_values, log(r_bar_values[i,:]) / log(10.0), 'k-', label=r'$log_{10}(\xi) = $%5.3f' % xi, linewidth = 5.0 * (i + 1.0) / (xi_num + 1.0))
plt.ylabel(r'$\log_{10}\left( \frac{R_3}{n^2 \bar{v} b^5} \right)$', fontsize=20)
plt.xlabel(r'$\log_{10}(\eta)$', fontsize=14)
plt.xlim([-2.0,1.0])
plt.ylim([-8.0,1.0])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='best')
# ax.set_aspect(0.50)
ax.minorticks_on()
ax.tick_params(which='major', length=10, width=2, direction='in')
ax.tick_params(which='minor', length=5, width=2, direction='in')
ax.xaxis.set_ticks_position('both')
ax.yaxis.set_ticks_position('both')
```

```python
xi_min = -1.0
xi_max = 3.0
xi_num = 100

eta_min = -2.0
eta_max = 1.0
eta_num = 100

xi_values = linspace(xi_min, xi_max, num=xi_num)
eta_values = linspace(eta_min, eta_max, num=eta_num)
r_bar_values = zeros((xi_num, eta_num))

for i, xi in enumerate(xi_values):
    for j, eta in enumerate(eta_values):
        r_bar_values[i,j] = log(r_bar(10**xi, 10**eta)) / log(10.0)
```

```python
origin = 'lower'

fig1, ax2 = plt.subplots(constrained_layout=True)
CS = ax2.contourf(eta_values, xi_values, r_bar_values, 100, cmap=plt.cm.bone, origin=origin)

CS2 = ax2.contour(CS, levels=CS.levels[::20], colors='r', origin=origin)

ax2.set_title(r'$R_3(\xi,\eta)$', fontsize=18)
ax2.set_xlabel(r'$\log_{10}(\eta)$', fontsize=14)
ax2.set_ylabel(r'$\log_{10}(\xi)$', fontsize=14)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
ax2.tick_params(which='major', length=7, width=2, direction='in')
ax2.xaxis.set_ticks_position('both')
ax2.yaxis.set_ticks_position('both')

cbar = fig1.colorbar(CS)
cbar.ax.set_ylabel(r'$\log_{10}\left( \frac{R_3}{n^2 \bar{v} b^5} \right)$', fontsize=20)
cbar.add_lines(CS2)
```
